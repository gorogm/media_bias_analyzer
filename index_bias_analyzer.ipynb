```python
import json
import pandas as pd
```

```python
topics = []
for filename in ['index_belfold2.json', 'index_kulfold2.json', 'index_gazdasag2.json']:
    with open(filename) as json_file:
        data = json.load(json_file)
    topics.extend([str(f['title']) + ('. ' + str(f['ajanlo']) if f['ajanlo'] != None else '' ) for f in data])
```

```python
len(topics)
```

```python
def queryString(query):
    query = query.lower()
    return [f for f in topics if query in f.lower()]
```

```python
queryString('versenytárs')
```

```python
import nltk
```

```python
def mostCommon(query, n=100):
    lines = queryString(query)
    words = ' '.join(lines).lower().replace(query.lower(), '')
    words = words.replace('.', ' ').replace(':', ' ').replace('"', ' ').replace("'", ' ').replace(",", ' ')
    words = words.replace(" az ", ' ').replace(' a ', ' ').replace(' hogy ', ' ').replace(' és ', ' ').replace(' is ', ' ')
    words = words.replace(" nem ", ' ').replace(' egy ', ' ').replace(' de ', ' ').replace(' lesz ', ' ')
    words = words.replace(" ha ", ' ').replace(' van ', ' ').replace(" azt ", ' ').replace(' sem ', ' ')
    words = words.replace("  ", ' ').replace("  ", ' ')
    fdist1 = nltk.FreqDist(words.split(' '))
    return fdist1.most_common(n) 
    
mostCommon(' lmp ', 10)
```

```python
def diff(q1, q2, n=40):
    mostCommon1 = mostCommon(q1, n)
    mostCommon2 = mostCommon(q2, n)
    mostCommon1 = [f[0] for f in mostCommon1]
    mostCommon2 = [f[0] for f in mostCommon2]
    print(q1)
    print(list(set(mostCommon1) - set(mostCommon2)))
    print(q2)
    print(list(set(mostCommon2) - set(mostCommon1)))

diff('mszp', 'fidesz')
```
